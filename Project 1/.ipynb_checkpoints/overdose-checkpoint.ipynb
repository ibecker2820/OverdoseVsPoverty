```python
# Build a scatter plot for each data type
plt.scatter(prate_2015,orate_2015, marker="o")
x_values= prate_2015
y_values= orate_2015

# Incorporate the other graph properties
plt.title("Overdose rates vs Poverty rates (2015)")
plt.ylabel("Overdose Rates 2015")
plt.xlabel("Poverty Rates 2015")
# Perform a linear regression on Overdose Rates vs. Top States
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)
# Get regression values
regress_values = x_values * slope + intercept
print(regress_values)
# Print r square value
print(f"R squared: {rvalue**2}")
# Create line equation string
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(line_eq)
plt.plot(x_values,regress_values,"r-")
plt.grid(True)
# Save the figure
plt.savefig("ScatterplotOverdosevsPovertryRates2015.png")
# Show plot
plt.show()

```
